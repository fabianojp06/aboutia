```python
from groq import Groq

client = Groq()
completion = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
      {
        "role": "user",
        "content": ""
      }
    ],
    temperature=1,
    max_completion_tokens=8192,
    top_p=1,
    reasoning_effort="medium",
    stream=True,
    stop=None
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```

```python
import os
from dotenv import load_dotenv, find_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 1. Carrega as variáveis de ambiente (Chave)
load_dotenv(find_dotenv())

# Verificação de segurança
if not os.environ.get("GROQ_API_KEY"):
    raise ValueError("❌ Chave GROQ_API_KEY não encontrada no .env")

# 2. Inicializa o modelo (O ChatGroq já cria o cliente sozinho)
# Removida a linha 'client = Groq(...)' que estava causando o erro
model = ChatGroq(model="llama-3.3-70b-versatile")

# 3. Define o Prompt
prompt = ChatPromptTemplate.from_template(
    "Você é um especialista em IA. Explique o conceito de {tema} de forma técnica e concisa."
)

# 4. Cria a Chain (Cadeia)
chain = prompt | model | StrOutputParser()

# 5. Executa
try:
    print("⏳ Processando...")
    resposta = chain.invoke({"tema": "Embeddings"})
    print("\nRESPOSTA DA IA:")
    print(resposta)
except Exception as e:
    print(f"Erro: {e}")
```

```python
import os
from groq import Groq # Importação necessária
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())

# 1. Cria o Cliente (Igual ao OpenAI())
client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),
)

# 2. A Chamada (Mudança Importante: messages em vez de prompt)
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile", # O modelo da Groq
    messages=[
        # No formato Chat, precisamos definir quem está falando
        {"role": "user", "content": "Crie uma canção contendo apenas uma estrofe."}
    ],
    temperature=0.7 
)

# 3. Pegando a resposta
print(response.choices[0].message.content)
```

```python
import os
from langchain_groq import ChatGroq
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())

# 1. Instancia o Modelo
llm = ChatGroq(model="llama-3.3-70b-versatile")

# 2. Chamada Direta (O LangChain converte a string para o formato de chat automaticamente)
response = llm.invoke("Crie uma canção contendo apenas uma estrofe.")

# 3. Resposta
print(response.content)
```
